## Question 1:What are Type I and Type II Errors in Hypothesis Testing, and how do they impact decision-making?

### Ans:Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. In this process, we start with two competing hypotheses:

* **Null Hypothesis (H₀):** There is no effect or difference.
* **Alternative Hypothesis (H₁ or Ha):** There is an effect or difference.

Since these decisions are based on limited sample data, errors are possible. These are known as **Type I and Type II errors**.

### **1. Type I Error (False Positive)**

* A **Type I error** occurs when we **reject the null hypothesis (H₀) even though it is true**.
* It means we conclude that a difference or effect exists when, in reality, it does not.
* The probability of making this error is denoted by **α (alpha)**, known as the **level of significance**.

#### **Example:**

Suppose a new medicine is being tested.

* **H₀:** The medicine has no effect.
* **H₁:** The medicine is effective.
  If researchers conclude the medicine works when it actually doesn’t, they have made a **Type I error**.

#### **Representation:**

Reality: H₀ is true
Decision: Reject H₀

### **2. Type II Error (False Negative)**

* A **Type II error** occurs when we **fail to reject the null hypothesis (H₀) even though it is false**.
* This means we miss detecting an effect that actually exists.
* The probability of this error is denoted by **β (beta)**.

#### **Example:**

Using the same medicine test:
If researchers conclude the medicine has no effect when it actually does, they have made a **Type II error**.

#### **Representation:**

 Reality: H₀ is false
 Decision: Fail to reject H₀

### **3. Relationship Between Errors and Power of a Test**

* The **power of a statistical test** is the probability of correctly rejecting a false null hypothesis.
* Power = 1 - β
* Increasing the sample size generally **reduces both α and β**, improving decision accuracy.


### **4. Impact on Decision-Making**

| **Error Type**    | **Meaning**                               | **Consequence**                 | **Real-life Example**         |
| ----------------- | ----------------------------------------- | ------------------------------- | ----------------------------- |
| **Type I Error**  | Rejecting a true null hypothesis          | False alarm or overreaction     | Approving an ineffective drug |
| **Type II Error** | Failing to reject a false null hypothesis | Missed opportunity or ignorance | Rejecting a useful drug       |

#### **Impact Summary:**

* **Type I error** → Leads to **unnecessary actions or wasted resources**.
* **Type II error** → Leads to **missed detection of real issues**.
* Decision-makers must **balance α and β** based on the seriousness of each error in their specific context. For example:

  * In medical research, **Type I error** is more dangerous (false claim of cure).
  * In security systems, **Type II error** is more dangerous (failing to detect threats).

### **Conclusion**

Type I and Type II errors are unavoidable risks in hypothesis testing.
Effective decision-making requires minimizing these errors through appropriate **sample size**, **significance level selection**, and **power analysis**.
A good researcher always aims to strike a **balance between accuracy and practicality** while interpreting hypothesis testing results.

### Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

### Answer:The **p-value** (probability value) measures the **likelihood of obtaining the observed results, or more extreme results, if the null hypothesis (H₀) is true**.

* A **small p-value (≤ α)** indicates **strong evidence against H₀**, so we **reject H₀**.
* A **large p-value (> α)** means the data is **consistent with H₀**, so we **fail to reject H₀**.

**Example:** If p = 0.03 and α = 0.05 → reject H₀ (result is statistically significant).
If p = 0.25 → fail to reject H₀ (no significant evidence).

In short, the **smaller the p-value, the stronger the evidence against the null hypothesis.**

### Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

### Answer:Both **Z-test** and **T-test** are statistical methods used to compare sample means and test hypotheses, but they differ mainly in sample size and knowledge of population variance.

| **Basis**                    | **Z-Test**                                      | **T-Test**                                        |
| ---------------------------- | ----------------------------------------------- | ------------------------------------------------- |
| **Population variance (σ²)** | Known                                           | Unknown                                           |
| **Sample size (n)**          | Large (n ≥ 30)                                  | Small (n < 30)                                    |
| **Distribution used**        | Standard Normal (Z)                             | Student’s t-distribution                          |
| **Shape**                    | Fixed                                           | Changes with sample size (flatter for small n)    |
| **Use case**                 | When population SD is known and sample is large | When population SD is unknown and sample is small |


### **When to Use:**

* **Z-Test:** Comparing large sample means or proportions when population variance is known.
* **T-Test:** Comparing small sample means when population variance is unknown.

**Example:**

* Z-test → Testing average height of 200 students.
* T-test → Testing average marks of 15 students in a class.

**In short:**
Use **Z-test** for large samples with known population SD, and **T-test** for small samples with unknown population SD.

### Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

### Answer:A **confidence interval (CI)** is a **range of values** that estimates an unknown population parameter (like the mean) with a certain level of confidence — usually **95% or 99%**.

[
\text{Confidence Interval} = \text{Sample Statistic} \pm \text{Margin of Error}
]

* The **margin of error (ME)** represents the **maximum expected difference** between the sample estimate and the true population value.
* A **larger margin of error** → **wider confidence interval** → **less precise estimate**.
* A **smaller margin of error** → **narrower confidence interval** → **more precise estimate**.

### **Example:**

If the mean income = ₹50,000 and the 95% CI is (₹48,000, ₹52,000),
we are **95% confident** that the true average income lies between ₹48,000 and ₹52,000.

**In short:**
The **confidence interval** shows the reliability of an estimate, and the **margin of error** controls how wide or narrow that estimate range is.

### Question 5:Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

### Answer:**ANOVA (Analysis of Variance)** is a statistical test used to **compare the means of three or more groups** to determine if at least one group mean is significantly different from the others.

### **Purpose:**

To test whether **differences among group means** are due to real effects or just random variation.

* **Null hypothesis (H₀):** All group means are equal.
* **Alternative hypothesis (H₁):** At least one mean is different.

### **Assumptions:**

1. **Independence:** Samples are independent of each other.
2. **Normality:** Data in each group follows a normal distribution.
3. **Homogeneity of variances:** All groups have equal variances.


### **Extension Beyond Two Groups:**

* A **t-test** compares **two means**, but **ANOVA** compares **three or more means simultaneously**, reducing the risk of Type I error.
* It uses the **F-statistic** to decide whether to reject H₀.

### **In short:**

**ANOVA** helps identify whether mean differences among multiple groups are statistically significant, under the assumptions of normality, equal variance, and independent samples.

## Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

```python
import numpy as np
from scipy import stats

data = [72, 75, 78, 71, 69, 74, 77, 73, 70, 76]
pop_mean, pop_std = 70, 5
n = len(data)

z = (np.mean(data) - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z = {z:.2f}, p = {p:.4f}")
print("Reject H₀" if p < 0.05 else "Fail to reject H₀")
```

## Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data from an exponential (non-normal) distribution
data = np.random.exponential(scale=2, size=10000)

# Apply Central Limit Theorem
sample_means = [np.mean(np.random.choice(data, 50)) for _ in range(1000)]

# Plot original vs sampling distribution
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title("Original Non-Normal Distribution")

plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title("Sampling Distribution (Approx. Normal)")

plt.show()
```

## Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error
    margin = sem * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    lower, upper = mean - margin, mean + margin

    # Visualization
    plt.errorbar(1, mean, yerr=margin, fmt='o', color='red', capsize=5)
    plt.title(f"{int(confidence*100)}% Confidence Interval")
    plt.ylabel("Sample Mean")
    plt.xticks([])
    plt.show()

    return lower, upper

# Example
data = np.random.normal(50, 10, 100)
ci = confidence_interval(data)
print("Confidence Interval:", ci)
```

## Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

```python
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [25, 30, 20, 25]
expected = [25, 25, 25, 25]

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic = {chi_stat:.2f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject H₀: Observed distribution differs significantly from expected.")
else:
    print("Fail to reject H₀: No significant difference between observed and expected.")
```

### **Interpretation**-The Chi-square goodness-of-fit test checks whether the observed data matches the expected theoretical distribution. Since p-value > 0.05, we fail to reject the null hypothesis, meaning the observed frequencies fit the expected distribution well.
